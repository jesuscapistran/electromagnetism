```python
# Librerias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Constantes 

Ke = 8.98e9 #[Nm^2/C^2] Cte de Coulomb en Sistema Internacional 
a = 1e-3  # 1 mm de radio alambre
b = 1e-2  # 1 cm de grosor del cable 
l = 1     # 1 m de longitud de cable coaxial  
```

```python
# Calculo de capacitancia

c = l/(2*Ke*np.log(b/a))
c = c*(1/1e-12) # picoFaradas 1e-12
round(c,3)
```

```python
# ¿Como Aumentamos la capacitancia? 
# Aumentamos radio interno del alambre de cobre 

a2 = np.arange(1e-3, 1e-2, 1e-4)
c2 = l/(2*Ke*np.log(b/a2))
c2 = c2*(1/1e-12) # picoFaradas 1e-12


plt.plot(a2/1e-3,c2)
plt.plot(a/1e-3,c,'k*')
plt.xlim(0.7,8)
plt.ylim(0,200)
plt.xlabel('a [mm]')
plt.ylabel('C [pF]')
plt.show
```

```python
# ¿Como Aumentamos la capacitancia? 
# Aumentamos la longitud del cable 

l3 = np.arange(1, 22, 0.5)
c3 = l3/(2*Ke*np.log(b/a))
c3 = c3*(1/1e-12) # picoFarads


plt.plot(l3,c3)
plt.plot(l,c,'k*')
plt.xlim(0.5,10)
plt.ylim(0,200)
plt.xlabel('l [m]')
plt.ylabel('C [pF]')
plt.show
```
